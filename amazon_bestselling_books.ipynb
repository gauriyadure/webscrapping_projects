```python
names,authors,ratings,rev,price=[],[],[],[],[]
for i in range(1,3):
    
    url=f"https://www.amazon.in/gp/bestsellers/books/ref=zg_bs_pg_{i}?ie=UTF8&pg={i}"
    page=requests.get(url)
    soup=bsp(page.content,'lxml')
    all_books=soup.find_all(class_="zg-item-immersion")
    for f in all_books:
        if f.find(class_="a-icon-alt"):
            if f.find(class_="a-size-small a-link-child"):
                names.append(f.find(class_="p13n-sc-truncate p13n-sc-line-clamp-1").get_text().strip())
                authors.append(f.find(class_="a-size-small a-link-child").get_text())
                ratings.append(f.find(class_="a-icon-alt").get_text())
                rev.append(f.find(class_="a-size-small a-link-normal").get_text().replace(',',''))
                price.append(f.find(class_="p13n-sc-price").get_text())

            elif f.find(class_="a-row a-size-small"):
                names.append(f.find(class_="p13n-sc-truncate p13n-sc-line-clamp-1").get_text().strip())
                authors.append(f.find(class_="a-row a-size-small").get_text())
                ratings.append(f.find(class_="a-icon-alt").get_text())
                rev.append(f.find(class_="a-size-small a-link-normal").get_text().replace(',',''))
                price.append(f.find(class_="p13n-sc-price").get_text())



        else:
            names.append(f.find(class_="p13n-sc-truncate p13n-sc-line-clamp-1").get_text().strip())
            authors.append(f.find(class_="a-row a-size-small").get_text())
            ratings.append(None)
            rev.append(None)
            price.append(f.find(class_="p13n-sc-price").get_text())

```

```python
d={'Name':names,'Author':authors,'Rating':ratings,'Reviews':rev,'Price':price}
d
```
