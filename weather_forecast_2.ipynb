```python
import requests
from bs4 import BeautifulSoup
url="https://forecast.weather.gov/MapClick.php?lat=40.41748000000007&lon=-86.27088999999995#.YasgXNBBzIU"
page=requests.get(url)
page.content
```

```python
soup=BeautifulSoup(page.content,'html.parser')
soup
```

```python
all_data=soup.find_all(class_="tombstone-container")
all_data
```

```python
len(all_data)
```

```python
first=all_data[0]
first
```

```python
pn=first.find(class_="period-name").get_text()
pn
```

```python
sd=first.find(class_="short-desc").get_text()
sd
```

```python
tm=first.find(class_="temp").get_text()
tm
```

```python
names,descs,temps=[],[],[]
for f in all_data:
    
    names.append(f.find(class_="period-name").get_text())
    descs.append(f.find(class_="short-desc").get_text())
    temps.append(f.find(class_="temp").get_text())
d={"Period Name":names,"Short descs":descs,"Temperature":temps}
d
```

```python
import pandas as pd
df=pd.DataFrame(d)
df
```
